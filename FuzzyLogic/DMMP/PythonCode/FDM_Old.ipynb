Angular Resolution Membership Functions: Illustrate the classification of observations into low, medium, and high categories based on the angular resolution of the observation in milli-arcseconds. This differentiation is crucial for detecting the fine details in gravitational lensing that may indicate the presence of fuzzy dark matter.

Density Membership Functions: Depict how the density of the gravitational lens, represented in arbitrary units, is categorised. The density reflects the mass distribution within the lens galaxy and influences the observed distortion patterns, which are key to inferring the properties of dark matter.

Width and Distribution Membership Functions: Show the classification of the observed gravitational arcs' width and distribution. These characteristics provide insights into the structure and distribution of dark matter in the lensing galaxy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate triangular membership value
def triangular_membership(x, a, b, c):
    if a <= x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    else:
        return 0

# Universe of discourse for each variable
x_resolution = np.arange(0, 11, 1)  # Angular resolution (0 to 10 milli-arcseconds)
x_density = np.arange(0, 101, 1)    # Density (0 to 100 arbitrary units)
x_width = np.arange(0, 51, 1)       # Width and distribution (0 to 50 arbitrary units)
x_mass = np.arange(0, 31, 1)        # Dark matter particle mass (0 to 30 eV, for illustration)

# Mock raw data
raw_resolution = 3  # milli-arcseconds
raw_density = 75    # arbitrary units
raw_width = 30      # arbitrary units

# Calculate membership values using the triangular membership function
resolution_low_degree = triangular_membership(raw_resolution, 0, 0, 5)
resolution_high_degree = triangular_membership(raw_resolution, 5, 10, 10)

density_low_degree = triangular_membership(raw_density, 0, 0, 50)
density_high_degree = triangular_membership(raw_density, 50, 100, 100)

width_narrow_degree = triangular_membership(raw_width, 0, 0, 25)
width_wide_degree = triangular_membership(raw_width, 25, 50, 50)

# Display the calculated membership degrees
print(f"Angular Resolution - Low: {resolution_low_degree:.2f}, High: {resolution_high_degree:.2f}")
print(f"Density - Low: {density_low_degree:.2f}, High: {density_high_degree:.2f}")
print(f"Width - Narrow: {width_narrow_degree:.2f}, Wide: {width_wide_degree:.2f}")
```

1. Sensitivity Analysis
This analysis will vary each input variable within its range while keeping others constant, to see how the output (estimated mass of dark matter particles) changes.

**Output Notes**

The sensitivity analysis plots this have generated below, show linear relationships between each input variable (angular resolution, density, and width) and the estimated mass of dark matter particles. This indicates that the simplistic formula used in this function treats the relationship between input variables and output as linear and directly proportional. In a real-world scenario, especially in a complex system like dark matter particle estimation, the relationships are unlikely to be linear.

The plots suggest that as each input variable increases, the estimated mass of dark matter particles also increases at a constant rate. The key points to consider here are:

Linearity: The relationships shown in the plots are linear, which may not capture the complexity of actual physical phenomena.

Sensitivity: The plots suggest that the output is uniformly sensitive to changes in input variables. In practice, certain variables may have a more pronounced or more subtle effect on the estimated mass.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a simplistic function to simulate FLS output based on input variables
# Note: In a real system, this would be replaced with your FLS inference engine
def fls_simulator(angular_resolution, density, width):
    # Simplified formula to estimate dark matter particle mass
    return (angular_resolution * 0.2 + density * 0.5 + width * 0.3) / 3

# Define the range for each variable
angular_resolution_range = np.linspace(0, 10, 50)
density_range = np.linspace(0, 100, 50)
width_range = np.linspace(0, 50, 50)

# Perform sensitivity analysis for each input variable
for variable, range_ in zip(['Angular Resolution', 'Density', 'Width'],
                            [angular_resolution_range, density_range, width_range]):
    mass_estimates = []
    for value in range_:
        if variable == 'Angular Resolution':
            mass_estimates.append(fls_simulator(value, 50, 25))  # Keeping others constant
        elif variable == 'Density':
            mass_estimates.append(fls_simulator(5, value, 25))
        else:  # Width
            mass_estimates.append(fls_simulator(5, 50, value))

    plt.figure()
    plt.plot(range_, mass_estimates)
    plt.title(f'Sensitivity Analysis - {variable}')
    plt.xlabel(variable)
    plt.ylabel('Estimated Dark Matter Particle Mass')
    plt.grid(True)
    plt.show()
```

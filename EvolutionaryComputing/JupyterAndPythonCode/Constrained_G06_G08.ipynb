```python
import numpy as np
import matplotlib.pyplot as plt

def g06(x):
    # Objective function for g06 problem
    return (x[0] - 10)**3 + (x[1] - 20)**3

def g06_constraints(x):
    # Constraints for g06 problem
    c1 = -(x[0] - 5)**2 - (x[1] - 5)**2 + 100
    c2 = (x[0] - 6)**2 + (x[1] - 5)**2 - 82.81
    return np.array([c1, c2]), []

def g08(x):
    # Objective function for g08 problem
    numerator = np.sin(2 * np.pi * x[0])**3 * np.sin(2 * np.pi * x[1])
    denominator = x[0]**3 * (x[0] + x[1])

    # Handle division by zero
    denominator[denominator == 0] = np.inf

    return -numerator / denominator

# Visualising g06 function
x = np.linspace(0, 30, 100)
y = np.linspace(0, 30, 100)
X, Y = np.meshgrid(x, y)
Z = g06([X, Y])
plt.contour(X, Y, Z, levels=50)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('g06 Objective Function')
plt.colorbar(label='Objective Function Value')
plt.show()

# Visualising g08 function
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = g08([X, Y])
plt.contour(X, Y, Z, levels=50)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('g08 Objective Function')
plt.colorbar(label='Objective Function Value')
plt.show()
```
